```python
import pandas as pd
import os

grocery_sales = pd.read_csv("grocery_sales.csv")
def extract(store_data, extra_data):
    extra_df = pd.read_parquet(extra_data)
    merged_df = store_data.merge(extra_df, on = "index")
    return merged_df
merged_df = extract(grocery_sales, "extra_data.parquet")
```

```python
def transform(raw_data):
  raw_data['Date'] = pd.to_datetime(raw_data['Date'])
  raw_data['Month'] = raw_data['Date'].dt.month
  clean_data = raw_data[
      ['Store_ID', 'Month', 'Date', 'Dept', 'IsHoliday', 'Weekly_Sales', 'CPI', 'Unemployment']
  ]
  return clean_data
clean_data = transform(merged_df)
print("üìÑ Cleaned Data:")
print(clean_data.info())
```

```python
def avg_weekly_sales_per_month(clean_data):
    agg_data = clean_data.groupby('Month',as_index=False)['Weekly_Sales'].mean()
    return agg_data
agg_data = avg_weekly_sales_per_month(clean_data)
print("üìä Aggregated Data (Average Weekly Sales per Month):")
print(agg_data)
```

```python
def load(full_data, full_data_file_path, agg_data, agg_data_file_path):
    full_data.to_csv(full_data_file_path, index=False)
    agg_data.to_csv(agg_data_file_path, index=False)
clean_data_path = "clean_data.csv"
agg_data_path = "agg_data.csv"

# Call the load() function
load(clean_data, clean_data_path, agg_data, agg_data_path)
```

```python
def validation(file_path):
    exists = os.path.exists(file_path)
    if exists:
        print(f"‚úÖ File exists: {file_path}")
    else:
        print(f"‚ùå File not found: {file_path}")
    return exists
cleaned_path = "clean_data.csv"
aggregated_path = "agg_data.csv"
validation(cleaned_path)
validation(aggregated_path)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
# Plotting the weekly sales data
plt.figure(figsize=(10, 6))

# Line plot of Weekly Sales over time (sorted by Date/Month)
plt.plot(clean_data.groupby('Month')['Weekly_Sales'].mean(), marker='o', linestyle='-', color='b')

plt.title('Average Weekly Sales by Month')
plt.xlabel('Month')
plt.ylabel('Average Weekly Sales ($)')
plt.grid(True)
```

```python
holiday_sales = clean_data[clean_data['IsHoliday'] == 1]
non_holiday_sales = clean_data[clean_data['IsHoliday'] == 0]

# Plot the weekly sales for holidays and non-holidays
plt.figure(figsize=(10,6))

# Plot Holiday Sales
plt.plot(holiday_sales['Date'], holiday_sales['Weekly_Sales'], label='Holiday Sales', color='red', alpha=0.7)

# Plot Non-Holiday Sales
plt.plot(non_holiday_sales['Date'], non_holiday_sales['Weekly_Sales'], label='Normal Day Sales', color='blue', alpha=0.7)
# Title and labels
plt.title("Weekly Sales: Holiday vs Normal Days")
plt.xlabel("Date")
plt.ylabel("Weekly Sales")
plt.legend()

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```
